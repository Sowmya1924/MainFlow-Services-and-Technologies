# Data Visualization

```python
plt.figure(figsize=(8, 6))
df['bedrooms'].value_counts().plot(kind='bar')
plt.title('number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.scatter(df.price,df.sqft_living,color='g')
plt.xlabel('sqft')
plt.ylabel('price')
plt.show()
```

```python
df.hist(figsize=(15,12))
plt.grid(False)
plt.show()
```
