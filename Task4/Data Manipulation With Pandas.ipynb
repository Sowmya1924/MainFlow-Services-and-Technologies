# Filtering Using Conditionals

```python
data=(df['Sex']=='female')
```

```python
data
```

```python
df.loc[data, 'Age']
```

```python
df.loc[df['Age']<19.0]
```

```python
df.tail()
```

Filter Multiple Values

```python
Survive=(df['Survived']>0)
```

```python
df.loc[Survive]
```

```python
df.loc[Survive, ['Name','Age','Sex']]
```

```python
names=['Najib, Miss. Adele Kiamie "Jane"','Graham, Miss. Margaret Edith','Behr, Mr. Karl Howell']
```
